```python
from langchain.embeddings import HuggingFaceEmbeddings

# Utiliser un modle efficace
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Diviser le document en petits textes (chunks)
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

texts = text_splitter.split_documents(documents)
```

```python
from langchain.vectorstores import FAISS

# Crer lindex FAISS  partir des textes et des embeddings
vectorstore = FAISS.from_documents(texts, embedding)
```

```python
# Liste de questions
queries = [
    "Quels sont les dbouchs du master WISD ?",
    "Quels sont les objectifs principaux du master WISD ?"

]

# Boucle sur chaque question
for q_idx, query in enumerate(queries, 1):
    print(f"\n Question {q_idx}: {query}\n")

    docs_similaires = vectorstore.similarity_search_with_score(query, k=3)

    for i, (doc, score) in enumerate(docs_similaires, 1):
        similarity_percent = (1 / (1 + score)) * 100
        print(f" Rsultat {i} (similarit  {similarity_percent:.2f}%) :\n{doc.page_content}\n{'-'*80}")
```

```python
from langchain.vectorstores import FAISS  # ou ton vectorstore
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain.chat_models import ChatOpenAI
from langchain.schema import Document

from groq import Groq
from langchain.llms.base import LLM

# 1. Adapter Groq en LLM LangChain

class GroqLLM(LLM):
    def __init__(self, api_key: str, model_name: str = "llama3-8b-8192"):
        self.client = Groq(api_key=api_key)
        self.model_name = model_name

    @property
    def _llm_type(self) -> str:
        return "groq"

    def _call(self, prompt: str, stop=None) -> str:
        messages = [{"role": "user", "content": prompt}]
        response = self.client.chat.completions.create(
            model=self.model_name,
            messages=messages
        )
        return response.choices[0].message.content

# 2. Ton PromptTemplate dj prt (exemple)

template = """
Tu es un assistant expert. En te basant uniquement sur le contexte suivant, rponds clairement  la question.

Contexte:
{context}

Question:
{question}

Rponse:
"""

prompt = PromptTemplate(input_variables=["context", "question"], template=template)

# 3. Instanciation LLM Groq

groq_llm = GroqLLM(api_key="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")

# 4. Chargement de ta vectorstore (exemple FAISS dj cr)
# vectorstore = FAISS.load_local("chemin_vers_index")  #  adapter

# 5. Cration de la chane RetrievalQA avec ton prompt

qa_chain = RetrievalQA.from_chain_type(
    llm=groq_llm,
    chain_type="stuff",  # "stuff" injecte tout le contexte dans le prompt, tu peux choisir "map_reduce" si gros contexte
    retriever=vectorstore.as_retriever(search_kwargs={"k": 3}),
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt}
)

# 6. Liste de questions  poser

queries = [
    "Quels sont les dbouchs du master WISD ?",
    "Quels sont les objectifs principaux du master WISD ?"
]

# 7. Boucle pour interroger avec multi-query retrieval

for i, question in enumerate(queries, 1):
    print(f"\n Question {i}: {question}\n")
    result = qa_chain(question)

    print(" Rponse :")
    print(result['result'])
    print("\n Documents sources :")
    for doc in result['source_documents']:
        print(f"- {doc.metadata.get('source', 'source inconnue')}")
        print(doc.page_content[:300] + "...\n")
    print("="*80)
```
